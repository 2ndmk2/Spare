```python

import pandas as pd
from scipy.interpolate import UnivariateSpline
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
import matplotlib
%load_ext autoreload
%autoreload 2
%matplotlib inline

import sys
import os
sys.path.append(os.getcwd()+"/../Spare_module/")
print(os.getcwd()+"/../Spare_module/")
from tsv_healpix_updated import *

```

```python
## Geometric parameters 
inc = 0 
Thetaeq=np.pi
zeta=90 * np.pi/180.0                                                                                        
Pspin=23.9344699/24.0                                                                                                                                                                                                           
wspin=2*np.pi/Pspin                                                                                                          
Porb=365.242190402                                            
worb=2*np.pi/Porb                                                                                                                
N=1024


### Calculation of Kernel Function & Light Curve
file_data = "../Data_mock/mockalbedo16.fits"
nside=16
mmap=hp.read_map(file_data)
obst=np.linspace(0.0,Porb,N)

Thetav=worb*obst
Phiv=np.mod(wspin*obst,2*np.pi)
WI,WV=comp_weight(nside,zeta,inc,Thetaeq,Thetav,Phiv)
W=WV[:,:]*WI[:,:]
W_mean = np.mean(W,axis = 0)

def best_map_return(folder_tsv, l1_min, l1_max, ltsv_min, ltsv_max, step,  land_f_fan, sn_100=False):

    l1_arr =  np.arange(l1_min, l1_max+step, step)
    ltsv_arr =  np.arange(ltsv_min, ltsv_max+step, step)
    red = []
    red2 = []
    red_min = 1e10
    red_min2= 1e10

    for l1_now in l1_arr:
        for ltsv_now in ltsv_arr:
            data = folder_tsv + "l1_%.2f_ltsv_%.2f.npy" % (l1_now, ltsv_now)
            if sn_100:
                data = folder_tsv + "l1%.2f_tsv%.2f.npy" % (l1_now, ltsv_now)
            mmap_l1_tsv = np.load(data) 
            red.append(np.sum(W_mean * (mmap_l1_tsv-land_f_fan)**2))
            red2.append(np.sum((mmap_l1_tsv-land_f_fan)**2))
            if red_min > np.sum(W_mean * (mmap_l1_tsv-land_f_fan)**2):
                red_min = np.sum(W_mean * (mmap_l1_tsv-land_f_fan)**2)
                data_best_l1_tsv = mmap_l1_tsv
                best_name_l1_tsv = folder_tsv + "l1%.2f_ltsv%.2f.npy" % (l1_now, ltsv_now)
    return data_best_l1_tsv, best_name_l1_tsv, red_min


import glob 

def return_match_map(folder_map, input_map):
    files = glob.glob(folder_map + "model_*.npy")
    for file in files:
        map_now = np.load(file)
        dif = np.sum(np.abs(map_now - input_map))
        if dif<0.00000001:

            return 0

```

```python
folder_s2 = "./result/non_cv_for_sn_zeta90_teq180_inc0_sn_2/model/"
best_s2, best_name_s2, red_min_s2 = best_map_return(folder_s2, -3, 1, -3 ,1, 0.25, mmap)
print(best_name_s2)
tik_s2_map = np.load(folder_s2 + "best_tik.npy")
return_match_map(folder_s2, tik_s2_map)
folder_s5 = "/Users/masatakaaizawa/paper_mischellous/mapping_github/Mock_sec3/result_wo_cv_for_sn_9_19/non_cv_for_sn_zeta90_teq180_inc0_sn_5/model/"
best_s5, best_name_s5, red_min_s5 = best_map_return(folder_s5, -3, 1, -3 ,1, 0.25, mmap)
print(best_name_s5)
tik_s5_map = np.load(folder_s5 + "best_tik.npy")
return_match_map(folder_s5, tik_s5_map)

folder_s100 = "./result_wo_cv_for_sn_9_19/non_cv_for_sn_zeta90_teq180_inc0_sn_100/model/best_model_"
best_s100, best_name_s100, red_min_s100 = best_map_return(folder_s100, -1, 3, -1 ,3, 0.25, mmap, sn_100= True)
print(best_name_s100)
folder_s100 = "./result_wo_cv_for_sn_9_19/non_cv_for_sn_zeta90_teq180_inc0_sn_100/model/"
tik_s100_map = np.load(folder_s100 + "best_tik.npy")
return_match_map(folder_s100, tik_s100_map)
```

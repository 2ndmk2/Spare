## Setting geometry parameters

See https://arxiv.org/abs/1204.3504 in detail

```python
inc=np.pi/6.0
Thetaeq=np.pi
zeta=np.pi/2.0                                                                                               
Pspin=23.9344699/24.0                                                                                                                                                                                                           
wspin=2*np.pi/Pspin                                                                                                          
Porb=365.242190402                                            
worb=2*np.pi/Porb                                                                                                                
N=1024
s_n =5

obst=np.linspace(0.0,Porb,N)
```

### Calculation of Kernel Function & Light Curve

```python
### Calculation of Kernel Function & Light Curve
Thetav=worb*obst
Phiv=np.mod(wspin*obst,2*np.pi)
WI,WV=comp_weight(nside,zeta,inc,Thetaeq,Thetav,Phiv)
W=WV[:,:]*WI[:,:]



if not os.path.exists(folder_others + "light_theta.npy"):
    lc_original=np.dot(W,mmap)
    lc_mean = np.mean(lc_original)
    normalize_lc = (s_n**2)/lc_mean
    lc = np.random.poisson(lc_original * normalize_lc)/normalize_lc
    sigma_arr = np.sqrt(lc_original * normalize_lc)/normalize_lc
    Theta_lc = np.array([Thetav,lc])
    #np.save(folder_others + "light_theta", Theta_lc)
else:
    Theta_lc = np.load(folder_others + "light_theta.npy")
lc = Theta_lc[1]

```

```python
W_mean = np.mean(W,axis = 0)
hp.mollview(np.log(W_mean), title="Cloud-subtracted Earth",flip="geo",cmap=plt.cm.bone,min=0,max=np.max(np.log(W_mean)))
hp.graticule(color="white");
plt.savefig(folder_fig + "answer.pdf", dpi=200, bbox_inches="tight")
plt.show()
```

## Sparse mapping
"mfista_func_healpix" is the main function with varying ( lambda_l1,lambda_tsv).  
We adopt (lambda_l1, lambda_tsv) = $(10^{-1}, 10^{0})$, which gave the optimal solution in the last CV run (use cv_run.py for that)

```python

I_init = np.zeros(len(mmap)) + np.mean(lc)
lambda_tsv = 10**1
lambda_l1 = 10**0## Too large lambda_l1 might fail to solve equations
L_init= 1

proxmap = soft_threshold_healpix

weight_matrix, test = calc_neighbor_weightmatrix(hp, nside)
I_est = mfista_func_healpix(I_init, lc, W, weight =weight_matrix, L_init = L_init, sigma =sigma_arr, 
                            prox_map = proxmap, eps = 1e-8, miniter = 3000,
                            eta = 1.05, lambda_l1 =lambda_l1 , lambda_tsv = lambda_tsv , maxiter=3500)
np.save(folder_model + "best_model", I_est)
hp.mollview(I_est , title="",flip="geo",cmap=plt.cm.bone,min=0,max=1.0)
hp.graticule(color="white")
plt.savefig(folder_fig + "l1_tsv_best.pdf", dpi = 200)
```

```python
file = open("test.dat","w")
print (len(I_est[I_est ==0]), file = file)
plt.hist(I_est, bins = 100)
plt.title("l1:%.2f, ltsv:%.2f, 'num(I)==0 is %d'" % (lambda_l1, lambda_tsv, len(I_est[I_est ==0])))
plt.show()
plt.close()
```

```python
def ComputeContours(P,Levels, file_save):
    
    NPix=hp.pixelfunc.get_map_size(P)
    NSide=hp.pixelfunc.npix2nside(NPix)
    
    Lon,Lat=hp.pixelfunc.pix2ang(NSide,np.arange(NPix,dtype=int),lonlat=True)
    Lon=(Lon+180)%360-180
    Cont=plt.tricontour(Lon,Lat,P.reshape(-1),levels=Levels,colors='k')
    plt.close()
    
    Cont_coll= Cont.collections[0].get_paths()
    hp.visufunc.mollview(P.reshape(-1),flip='geo',cmap='brg', title='(a)')
    hp.graticule()
    
    for iCont in range(len(Cont_coll)):
        hp.visufunc.projplot(Cont_coll[iCont].vertices[:,0],Cont_coll[iCont].vertices[:,1],lonlat=True,color='k')
    
    plt.savefig(file_save, dpi = 200)
    plt.show()
    plt.close()
I_best = np.load(folder_model + "best_model.npy")
ComputeContours(I_best, [np.percentile(I_est, 70)], file_save ="test.pdf")
```
